```python
import ProperEvalFunctions as pef
import EvalFunctions as ef
import pandas as pd
import pickle
```

## Expectation machine

```python
path, os = pef.get_pathes("20210417_output")
analysis_path = path+os.sep+"analysis"+os.sep
Data = pd.read_csv(analysis_path+"Data_history.csv", index_col = 0, low_memory=False)
strainplate = pd.read_csv(analysis_path+"Strainplate.csv", index_col = 0)
Data[Data.turnover_id.isnull() == False]
strain_dict = pickle.load(open("strain_dict.pkl", "rb"))
```

## Initialize Extra columns

```python
Data["expected_phenotype"] = None
Data["expectation_machine"] = None
Data["expected_phenotype"] = Data["expected_phenotype"].apply(lambda x: [0, 0, 0, 0])
Data["input_wells"] = None

```

## Collect input wells
- add all wells that infected respective well
- infection + turnover + transfer
- make list

```python
Data.apply(lambda x: pef.input_wells(x, Data), axis=1)
Data.head()
```

## Repair Phenostring 
- was broken after csv import

```python
tmp = Data.apply(lambda x: pef.repair_array(x, Data), axis = 1)
```

## Inverse dictionary and repair strain dictionary

```python
inv_strain_dict = {v: k for k, v in strain_dict.items()}
inv_strain_dict.update({"wt":1})
inv_strain_dict.update({"bl":0})
inv_strain_dict.update({"AB":15})
```

## Add Pheno_str to strainplate

```python
strainplate["strain_real_num"] = strainplate.apply(lambda x: pef.strain_to_pheno_num(inv_strain_dict, x["strain_real"]), axis=1) 
strainplate["strain_real_str"] = strainplate.apply(lambda x: pef.create_binary_array(x["strain_real_num"]), axis = 1)
strainplate.head()
```

## Add Expectations 

```python
Data["expected_pheno_str"] = None
Data.apply(lambda x: pef.expect_phenotype(x, Data, strainplate), axis = 1)
Data['expected_pheno_num'] = Data.apply(lambda x:  pef.strain_arr_to_num(x["expected_pheno_str"]) , axis=1)
Data["expected_phenotype"] = Data.apply(lambda x:  pef.strain_num_to_pheno(strain_dict, x["expected_pheno_num"]) , axis=1)
Data = Data.reindex(sorted(Data.columns), axis=1)
Data.head()
```

## Create Predictions only based on Expectations
- Clearence rate is 100% for both antibiotics
- A&B + treatment_A -> B 
- A&B 50% chance -> AB

```python
## Create Treatment_index_dict
treatment_index_dict={"none":3, "A" : 1, "B" : 2, "AB" : 0}

Data["expected_pheno_str_machine"] = None      
Data["treated_pheno_str_machine"] = None
Data["expected_pheno_str_machine"] = Data["expected_pheno_str_machine"].apply(lambda x: [0, 0, 0, 0]) 

for t in sorted(Data.transfer_n.unique()):
    Data_t = Data[Data.transfer_n == t]
    
    ## Create expected pheno_str by collecting the phenotypes from all wells - but this time collect machine values
    Data_t["expected_pheno_str_machine"]  = Data_t.apply(lambda x: pef.expect_phenotype_machine(x, Data, strainplate), axis = 1)
    Data_t = Data[Data.transfer_n == t]
    
    ## Conjugation events
    Data_t["expected_pheno_str_machine"]  = Data_t.apply(lambda x: pef.superconjugation(x["expected_pheno_str_machine"], 0.5), axis=1)
    
    ## Treat Phenostrings
    Data_t["treated_pheno_str_machine"] = Data_t.apply(lambda x: pef.treat_phenotype(x["expected_pheno_str_machine"], x["treatment_with"], treatment_index_dict), axis=1)
    
    ## Calculate machine pheno num
    Data_t['treated_pheno_num_machine'] = Data_t.apply(lambda x:  pef.strain_arr_to_num(x["treated_pheno_str_machine"]) , axis=1)
    
    ## Calculate machine phenotype
    Data_t["treated_phenotype_machine"] = Data_t.apply(lambda x:  pef.strain_num_to_pheno(strain_dict, x["treated_pheno_num_machine"]) , axis=1)
    

    ## Integrate Data_t in Data
    Data.loc[Data_t.index, Data_t.columns] = Data_t
    
    
    
Data_t[["expected_pheno_str_machine", "treated_pheno_str_machine", "treated_pheno_num_machine", "treated_phenotype_machine"]]

```
