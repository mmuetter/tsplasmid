```python
import pandas as pd
import EvalFunctions as ef
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.ticker as plticker
import ProperEvalFunctions as pef
import pickle
import VisuFunctions as vf
import StatFunctions as sf
```

```python
path, os = pef.get_pathes("20210417_output")
analysis_path = path+os.sep+"analysis"+os.sep
hue_infos, strategies, plate_rows, agar_strain_dict = pef.build_dictionaries()
strategies_rv = {v: k for k, v in strategies.items()}
```

```python
summary_machine = pd.read_csv(analysis_path+"summary_machine.csv", index_col = 0)
summary = pd.read_csv(analysis_path+"summary.csv", index_col = 0)
```

```python
summary_machine.head()
```

```python
## Create unique index for each Strain per strategy, replicate and time and engine
## For the expectation machines 
summary_machine["sum_id"] = summary_machine.apply(lambda x: sf.make_sum_id(x["transfer_n"], strategies_rv[x["strategy"]], x["treated_phenotype_machine"], x["replicate"], "m"), axis = 1)
try:
    summary_machine = summary_machine.set_index("sum_id")
except:
    pass
summary_machine = summary_machine.iloc[:,0:5]

## For the observations
summary["sum_id"] = summary.apply(lambda x: sf.make_sum_id(x["transfer_n"], strategies_rv[x["strategy"]], x["phenotype"], x["replicate"], "r"), axis = 1)
try:
    summary = summary.set_index("sum_id")
except:
    pass
summary = summary.iloc[:,0:5]

summary_machine.head()
```

```python
summary.head()
```

```python
## Make a difference between phenotypes weather they were observed or calculated
summary_machine["phenotype"] = summary_machine["treated_phenotype_machine"]#summary_machine.apply(lambda x: x["treated_phenotype_machine"]+"_m", axis=1)
summary_machine = summary_machine.drop(["treated_phenotype_machine"], axis=1)
```

```python
summary["style"] = "experiment"
summary_machine["style"] = "simulation"
full_summary = summary.append(summary_machine)
full_summary.head()
```

```python
sns.set_style("darkgrid")
g = sns.relplot(
    data=full_summary, x="transfer_n", y="fraction",
    col="strategy", col_wrap = 3,  hue="phenotype", 
    kind="line", style="style", 
    palette=hue_infos, ci=95
)

g.set_axis_labels("Transfer", "Fraction [%]")

# Adjust grid spacing
g.axes[0].yaxis.set_major_locator(plticker.MultipleLocator(base=10))
g.axes[0].xaxis.set_major_locator(plticker.MultipleLocator(base=2))

plt.savefig(analysis_path+'relplot_both.png', bbox_inches="tight", dpi=500)
```

```python
sns.set_style("darkgrid")
sns.set(font_scale=1.5) 

g = sns.relplot(
    data=full_summary, x="transfer_n", y="fraction",
    col="strategy", col_wrap = 2,  hue="phenotype", 
    kind="line", style="style", 
    palette=hue_infos, ci=95, 
    aspect = 1.4
)

g.set_axis_labels("Transfers", "Fraction [%]")

# Adjust grid spacing
g.axes[0].yaxis.set_major_locator(plticker.MultipleLocator(base=10))
g.axes[0].xaxis.set_major_locator(plticker.MultipleLocator(base=2))

g._legend.set_bbox_to_anchor([0.85, 0.83])

plt.savefig(analysis_path+'relplot_2_both.png', bbox_inches="tight", dpi=500)
```
