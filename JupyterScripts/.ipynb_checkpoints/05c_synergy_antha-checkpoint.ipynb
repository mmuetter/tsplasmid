```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import pickle
import numpy as np
from evaluation_scripts.base import get_pathes
from evaluation_scripts.well_ident import build_dictionaries
from model_scripts.simple_model import create_x_form_dict, treatment, make_T
jupyter_path, base_path, path, analysis_path, os = get_pathes("20220412_output")
hue_infos, strategy_dict, plate_rows, agar_strain_dict = build_dictionaries()
Data = pd.read_pickle(analysis_path+os.sep+"obj"+os.sep+"01b_Data.pkl")
dt = [0, max(Data.transfer_n)]
strategies = strategy_dict.values()
```

```python
transition_date = "av"## options: 1) experiment folder name, 2) "av" , 3) "native" (chose same matrix as from experiment)
use_av_transition_props = True
```

```python
import json
from pandas.io.json import json_normalize

with open(os.path.join(path, "exp_pars.json")) as json_data:
    exp_pars = json.load(json_data)

rates = exp_pars["rates"][0]
rates
```

```python
community = {}
X = ["UI", "S", "A_r", "B_r", "A&B", "AB_r"]
total = 0
for c in exp_pars["community"]:
    strain = c["strain"]
    if strain == "wt":
        strain = "S"
    n = c["n"]
    total += n 
    community.update({strain:n})

for key in community:
    community.update({key : community[key]/total})
community.update({"A&B":0})

reservoir = pd.DataFrame(columns = ["f"])
for strain in X:
    reservoir.loc[strain, "f"] = community[strain]
reservoir
```

```python
from model_scripts.pop_model import (
    conjugation, 
    netto_loss, 
    clearance, 
    make_cyc_treat, 
    find_rates, 
    make_timestep,
    map_results)



if transition_date == "av":
    R_path = os.path.join(base_path, "summary", "obj")
    
```

```python
from scipy.integrate import solve_ivp
model_data = pd.DataFrame(columns = ["t", "f", "phenotype", "strategy"])
for strategy in strategies:
    res = solve_ivp(make_timestep, dt, reservoir.f.values, args = (strategy, R_path, rates["infection"], rates["turnover"], reservoir.f.values, X))
    model_data = model_data.append(map_results(strategy, X, res), ignore_index = True)
model_data["F"] = model_data.f*100
model_data.to_pickle(os.path.join(analysis_path, "obj", "05b_pop_model_results.pkl"))
```
