```python
import pandas as pd
import EvalFunctions as ef
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import ProperEvalFunctions as pef
import pickle
import VisuFunctions as vf
import StatFunctions as sf
import math
import turnover_analysis_functions as taf
import numpy as np


path, os = pef.get_pathes("20210417_output")
analysis_path = path+os.sep+"analysis"+os.sep
Data = pd.read_pickle(analysis_path+"obj/Turnover_statistic.pkl").set_index("history_group")

Data.head(40)
```

```python
tau = 0.2
a = 1-tau
x0 = [1, 0, 0]
m0 = np.array([1, 0, 0])
m = tau * m0
def f_t (a, M, t, x0, m):
    f = np.matmul(a**t * np.linalg.matrix_power(M, t), x0 )
    for i in range(1,t+1):
        matrix = a**(i-1) * np.linalg.matrix_power(M, i)
        f = f+ np.matmul(matrix, m)
    return f
```

```python

clearance_matrix_exp = pd.DataFrame(columns = ["UI", "wt", "A_r", "B_r", "AuB", "AB"], 
                                    index = ["none", "Mono A", "Mono B", "Combo"],
                                    data = np.zeros([4,6]))
clearance_matrix_exp
```

```python
M = np.array([[0, 0, 0], [0, 0, 0], [0, 0, 1]]).astype(float)
for story, row in Data.iterrows():
    ## Get indices
    strings = story.split(" ")
    strain = strings[3]
    treatment = " ".join(strings[6:])
    
    ## Calculate stationary phase
    M[1,0] = Data.loc[story, "P(X, T=1)"]
    M[2,0] = 1-Data.loc[story, "P(X, T=1)"]
    M[1,1] = Data.loc[story, "P(X, T=2)"]
    M[2,1] = 1-Data.loc[story, "P(X, T=2)"]
    x_stat = tmp=f_t(a, M, 100, x0, m) 
    S_stat = x_stat[0] + x_stat[1]
    ms = m0[0] + m0[1]
    c_av = tau*(ms-S_stat)/S_stat
    if c_av < 0.0001:
        c_av
    ## Save value to clearance Matrix
    clearance_matrix_exp.loc[treatment, strain] = np.round(c_av, 2)
    
clearance_matrix_exp.to_pickle(analysis_path+"obj/clearance_matrix_exp.pkl")
clearance_matrix_exp
```

```python
T = np.linspace(0, 100,101)
y = []

for t in T:
    tmp=f_t(a, M, int(t), x0, m)
    y.append(tmp)
y = np.array(y)

import matplotlib.pyplot as plt
plt.plot(T, y, T, np.array(y)[:,0] + np.array(y)[:,1])
plt.legend(["S1", "S2", "UI", "S"])
```
