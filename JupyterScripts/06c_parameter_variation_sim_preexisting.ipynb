```python
%load_ext autoreload
%autoreload 2

from model_scripts.Parameter_Set_class import Parameter_set
from model_scripts.instrutcion_set_class import InstructionSet
from model_scripts.simulation_class import Simulation
import pandas as pd
from copy import deepcopy
from itertools import product
from evaluation_scripts.base import obj_to_pickle
import os
import numpy as np
```

```python
## Just for the shape (len(Transfer_n))
date = "20220127"
parameter = Parameter_set(date)

number_of_parameter_sets = 10000
number_of_var_sims = 10
```

```python
pars = {}
simulations = {}

for i in range(number_of_parameter_sets):    
    dfs = []
    parameter.randomize_parameter()
    print(i, parameter)
    instructions = InstructionSet(parameter=parameter)
    for j in range(number_of_var_sims):
        ## Simulation
        instructions.randomize_instructions()
        simulation = Simulation(instructions)
        simulation.run_simulation()
        
        df = simulation.timeplot_data
        df["rep"] = j
        dfs.append(df)
        
    ## Document
    p = deepcopy(parameter)
    par = {
        "turnover" :p.turnover,
        "infection":p.infection, 
        "community":p.community_distribution
    }    
    pars.update({i:par})
    simulations.update({i:pd.concat(dfs)})
    

    
```

```python
obj_to_pickle(os.path.join(parameter.pathes["base"], "sensitivity_sim", "sensitivity_sim_preexisting.pkl"), simulations)
obj_to_pickle(os.path.join(parameter.pathes["base"], "sensitivity_sim", "sensitivity_par_preexisting.pkl"), pars)
```
